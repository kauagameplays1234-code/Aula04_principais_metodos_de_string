```python
nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM "

# remover espaços
nome = nome.strip()
email = email.strip()

# deixar tudo minusculo
nome = nome.lower()
email = email.lower()

print(nome)
# substituir email
email = email.replace("empresa.com","gmail.com")

print(email)
# quatidade de caracteres
print('quantos caracteres tem no nome',len(nome))

# acessar elementos

print('primeira e ultima letra ',nome[0],nome[-1])
```
